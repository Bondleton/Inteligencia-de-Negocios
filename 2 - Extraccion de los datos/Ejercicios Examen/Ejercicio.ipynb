**Importar librerias**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Cargar archivo excel**

```python
df = pd.read_excel("Detalle precios y productos fabricados 2022.xlsx")
df
```

**Verificar información del DataFrame**

```python
df.info()
```

**Variables Cualitativas**

**Filtro 1 Por objeto**

```python
# Sintaxis filtro = df[(df["var"] == "var")]
filtro1 = df[(df["NOMBRE_VENDEDOR"] == "ALFREDO CANELA")]
filtro1
```

**Filtro 2: filtro Y**

Descripción: 
Filtrar por rango de valores
Selecciona los registros donde el PRECIO_UNITARIO esté entre 100 y 500.

```python
filtro2 = df[(df["PRECIO_UNITARIO"] > 100) & (df["PRECIO_UNITARIO"] < 500)]
filtro2
```

**Filtro 3: Por Comparación**

Descripción: Obtén todas las filas donde CANT sea mayor a 50.

```python
filtro3 = df[(df["CANT"] > 50)]
filtro3
```

**Filtro 4: Filtro Not**

Descipción: Filtra los registros donde la fecha (FECHA_DOC) no sea 28/01/2022 

```python
filtro4 =  df[~(df["FECHA_DOC"] == "2022-01-28")]
filtro4
```

**Filtro 5: Por filas consecutivas**

Descripción: Extrae las filas desde el índice 50 hasta el 100.

```python
filtro5 = df.iloc[ 50:101, :]
filtro5
```

**Filtro 6 Por columnas**

Descripción: Extrae solo las columnas CVE_DOC, NOMBRE_CLIENTE, y SUBTOTAL_PARTIDA.

```python
filtro6 = df.iloc[ : , [0,3,10]]
filtro6
```

**Filtro 7**

Descripción: Encuentra las filas donde NOMBRE_VENDEDOR sea nulo (NaN).

```python
filtro7 = df[(df["NOMBRE_VENDEDOR"] == "NaN")]
filtro7
```

**Filtro 8:**

Descripción: Filtra los registros donde PRECIO_UNITARIO sea mayor a 500 y CANT mayor a 10.

```python
# Sintaxis filtro = df[(df["var"] > cond) & ()]
filtro8 = df[(df["PRECIO_UNITARIO"] > 500) & (df["CANT"] > 10)]
filtro8
```

**Filtro 9**

Descripción: Obtén los registros donde NOMBRE_CLIENTE sea "ELICAMEX" o NOMBRE_VENDEDOR sea "ATENCION A CLIENTES".

```python
filtro9 = df[(df["NOMBRE_CLIENTE"] == "ELICAMEX") | (df["NOMBRE_VENDEDOR"] == "ATENCION A CLIENTES")]
filtro9
```

**Filtro 10**

Descripción: Encuentra todas las ventas realizadas después del "2022-06-01".

```python
filtro10 = df[(df["FECHA_DOC"] > "2022-06-01")]
filtro10
```
