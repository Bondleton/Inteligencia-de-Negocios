Determinacion de clases para datos agrupados

```python
# Corroborar valores nulos
valores_nulos = Micro_Retailer.isnull().sum()
valores_nulos
```

```python
# Calculamos el numero total de la poblacion "n"
Micro_Retailer['_latitude'].info()
n=171
```

```python
# Ontenemos el limite superior y el lim de la columna objetivo
Max = Micro_Retailer['_latitude'].max()
Min = Micro_Retailer['_latitude'].min()
Limites = [Min, Max]
Limites
```

```python
# Calculamos el rango R
R = Max - Min
R
```

```python
# Calculamos el numero de intervalos de clase "ni", aplicando la regla de Sturges
ni = 1+3.32*np.log10(n)
ni
```

```python
# Caluclamos el ancho del intervalo "i"
i = R/ni
i
```

Creacion de Categorias a partir de clases

```python
# Categorizacion de variables (min, max, cuantas clases voy a tener (8+1 clase mas)) Numero de intervalos
intervalos = np.linspace(18.9993095, 19.0752733, 9)
intervalos
```

```python
# Creamos las categorias
categorias = ["Categoria1","Categoria2", "Categoria3", "Categoria4", "Categoria5", "Categoria6", "Categoria7", "Categoria8"]
```

```python
# Finalmente creamos las categorias en la columna númerica
Micro_Retailer['_latitude'] = pd.cut(x = Micro_Retailer['_latitude'], bins=intervalos, labels = categorias)
Micro_Retailer['_latitude']
```
