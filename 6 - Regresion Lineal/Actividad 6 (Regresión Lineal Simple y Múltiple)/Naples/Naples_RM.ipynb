2. host_acceptance_rate

```python
from turtle import color

sns.scatterplot(x='review_scores_cleanliness', y='host_acceptance_rate', color="blue", data=Naples)
sns.scatterplot(x='host_is_superhost', y='host_acceptance_rate', color="red", data=Naples)
sns.scatterplot(x='property_type', y='host_acceptance_rate', color="orange", data=Naples)
```

```python
#Declaramos las variables dependientes e independientes para la regresión lineal
Vars_Indep= Naples[['review_scores_cleanliness', 'host_is_superhost','property_type']]
Var_Dep= Naples['host_acceptance_rate']
```

```python
#Se define model como la función de regresión lineal
from sklearn.linear_model import LinearRegression
model= LinearRegression()

#Verificamos la función relacionada al modelo
type(model)
```

```python
#Ajustamos el modelo con las variables antes declaradas
model.fit(X=Vars_Indep, y=Var_Dep)
```

```python
#Verificamos los coeficientes obtenidos para el modelo ajustado
model.__dict__
```

Modelo matemático = y -0.0137249 +  0.0196129 +  0.00018331 + 1.0303536041729697

```python
#Evaluamos la eficiencia del modelo obtenido por medio del coeficiente R Determinación
model.score(Vars_Indep,Var_Dep)
```

```python
#Predecimos los valores de total de accidentes a partir de las variables: "alcohol", "speeding" y "no_previous"
y_pred= model.predict(X=Naples[['review_scores_cleanliness', 'host_is_superhost', 'property_type']])
y_pred
```

```python
#Insertamos la columna de predicciones en el DataFrame
Naples.insert(26, 'Predicciones', y_pred)
Naples
```

```python
#Visualizamos la gráfica comparativa entre el total real y el total predecido
sns.scatterplot(x='review_scores_cleanliness', y='host_acceptance_rate', color="blue", data=Naples)
sns.scatterplot(x='review_scores_cleanliness', y='Predicciones', color="red", data=Naples)

# 'review_scores_cleanliness', 'host_is_superhost', 'property_type'

# sns.lineplot(x='review_scores_cleanliness', y='Predicciones', color="red", data=Naples)
```

```python
#Corroboramos cual es el coeficiente de Determinación de nuestro modelo
coef_Deter=model.score(X=Vars_Indep, y=Var_Dep)
coef_Deter
```

```python
#Corroboramos cual es el coeficiente de Correlación de nuestro modelo
coef_Correl=np.sqrt(coef_Deter)
coef_Correl
```
