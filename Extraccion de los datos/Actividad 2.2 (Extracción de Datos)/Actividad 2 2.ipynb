Importar librerias

```python
# Importar librerias 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Lectura del archivo Excel

```python
# Cargar el archivo sin indice
df= pd.read_excel('Datos de Facturación.xlsx')
df
```

Verificar información del DataFrame

```python
# Verificamos información del DataFrame
df.info()
```

Convetir de objetos a numerico

Descripción: aquí se convierte la variable CVE_CLPV de objeto a numero.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df['CVE_CLPV'] = le.fit_transform(df['CVE_CLPV'])

df.info()
```

**Filtro 1: Utilizando Filtro Y**

Descripción: 1-CVE_CLPV      (valores: 1000 a 2000) Con el filtro Y podemos buscar los valores que van de 1000 a 2000 de la variable CVE_CLPV.

```python
filtro1= df[(df["CVE_CLPV"] >= 1000) & (df["CVE_CLPV"] <= 2000)]
filtro1.to_csv("F1_Ejercicio1.csv")
filtro1
```

**Filtro 2: Utilizando filtro Not y filtro Y**

Descripción: 2-CVE_VEND     (Todas las claves excepto “5” y “4”) Con este filtro se juntam las negaciones de cuando buscamos los valores tanto para 4 como para 5.

```python
filtro2= df[~(df["CVE_VEND"] == 4) & ~(df["CVE_VEND"] == 5)]
filtro2.to_csv("F2_Ejercicio2.csv")
filtro2
```

**Filtro 3**

Descripción: 3-FECHA_ENT   (Fechas de “28/02/2022”) En este filtro buscamos la fecha 28/02/2022 en la variable Fecha_Ent, para este caso primero convertimos la fecha tipo objeto en tipo datetime.

```python
# Convertir la columna 'FECHA_ENT' a tipo datetime
df['FECHA_ENT'] = pd.to_datetime(df['FECHA_ENT'], errors='coerce')
df.info()
```

```python
# Ahora aplicar el filtro con la fecha correcta
filtro3 = df[(df["FECHA_ENT"] == "2022-02-28")]
filtro3.to_csv("F3_Ejercicio3.csv")
filtro3
```

**Filtro 4: Usando filtro O**

Descripción: 4-CAN_TOT (Cantidades menores a 5951.7)   o  STATUS (“E”) En este filtro buscamos que se cumplen una u otra de las condiciones establecidas.

```python
filtro4= df[(df["CAN_TOT"] < 5951.7 ) | (df["STATUS"] == "E")]
filtro4.to_csv("F4_Ejercicio4.csv")
filtro4
```

**Filtro 5: Filtro por columnas no consecutivas**


Descripción: 5- Solo las columnas: CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT
En este filtro se busca las columnas numero 0,6,7,9. 

```python
filtro5 = df.iloc[ : , [0,6,7,9]]
filtro5.to_csv("F5_Ejercicio5.csv")
filtro5
```

**Filtro 6: Por filas**

Descripción: 6-Solo las filas de 7001-7099
Aquí en este filtro se seleccionan filas consecutivas que van del 7001 al 7100 debido a que se excluye una por la posición.

```python
filtro6 = df.iloc[7001:7099, :]
filtro6.to_csv("F6_Ejercicio6.csv")
filtro6
```

**Filtro 7 :**

Descripción: 7-Index= CVE_VEND (valores: 1, 2)  (columna: FECHAELAB)
Se utiliza el parámetro index_col=3 para especificar que la primera columna del archivo será utilizada como índice del DataFrame.

```python
df1= pd.read_excel('Datos de Facturación.xlsx', index_col=3)
df1
```

Descripción: 7-Index= CVE_VEND (valores: 1, 2)  (columna: FECHAELAB), aquí filtramos con el indice de CVE_VEND y las filas especificas 1 y 2.

```python
filtro7 = df1.loc[[1,2],["FECHAELAB"]]
filtro7.to_csv("F7_Ejercicio7.csv")
filtro7
```
