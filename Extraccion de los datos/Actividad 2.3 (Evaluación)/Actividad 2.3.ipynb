**Ejercicio 1: usando filtro O**

Descripcion a) Los registros de 5 "host name" diferentes. Aqui se hace uso del filtro O, para que se filtren los datos de los 5 distintos Host.

```python
filtro1 = df[ (df["host_name"] == "Adelaide") | (df["host_name"] == "Achille") | (df["host_name"] == "Ada") | (df["host_name"] == "Adam") | (df["host_name"] == "Cebollitas") ]
filtro1.to_csv("Ejercicio1.csv")
filtro1
```

**Ejercicio 2: Usando el filtro por comparacion**

Descripción: b) Los registros de los host que se hallan unido a Airbnb "antes del año 2019" En este filtro es de por comparación, aquí estamos indicando que queremos las fechas antes de que inicie el año 2019.

```python
filtro2 = df[(df["host_since"] < "2018-12-31")]
filtro2.to_csv("Ejercicio2.csv")
filtro2
```

**Ejercicio 4: Por filtro Not**

c) Los registros de los host que responden "a más tardar en 1 día" (a few days or more) En este filtro not para solo seleccionar los datos donde se tardan en responder en menos de un día.

```python
filtro3=df[~(df["host_response_time"]  == "a few days or more")]
filtro3.to_csv("Ejercicio3.csv")
filtro3
```

**Ejercicio 4 : Por filtro de Comparación**

d) Los registros que tengan "al menos 5 accommodates" Aquí estamos filtrando por comparacion donde se seleccion los datos menos o iguales a 5 para la variable de accommodates

```python
filtro4 = df[(df["accommodates"] <= 5)]
filtro4.to_csv("Ejercicio4.csv")
filtro4
```

**Ejercicio 5: Por filtro O**

e) Los registros de los tipos de cuarto: "Entire home" y "Private room" En este ejercicio estamos haciendo uso del filtro O para que se cumpla almenos una de las condiciones.

```python
filtro5 = df[(df["room_type"] == "Entire home/apt") | (df["room_type"] == "Private room")]
filtro5.to_csv("Ejercicio5.csv")
filtro5
```

**Ejercicio 6: Usando filtro por comparación**

f) Los registros que "a lo mucho cuenten con 4 camas" Aqui se estan filtrando los datos que cumplan con la comparacion mientras sean camas de de la cantidad igual o menor a 4.

```python
filtro6 = df[(df["beds"] <= 4)]
filtro6.to_csv("Ejercicio6.csv")
filtro6
```

**Ejercicio 7 : Combinacion de filtros O y Y**

g) Los registros que "tengan 1 baño o que sean superhost" pero con la condición de que "todos sean instant_bookable" Aqui estamos usando un filtro combinado, primero se cumple el filtro O y despues el filtro Y, para que se cumpla la condicion de que tenga true en instant_bookable

```python
filtro7 = df[(df["bedrooms"] <= 1) | (df["host_is_superhost"] == "t") & (df["instant_bookable"] == "t")]
filtro7.to_csv("Ejercicio7.csv")
filtro7
```

**Filtro 8: Usando filtro Y**

h) Los registros cuyo "review_scores_cleanliness sea menor de 3 y review_scores_ rating sea mayor de 3" Aqui se esta haciendo uso del filtrado Y para que se cumplan ambas condidiones.


```python
filtro8= df[(df["review_scores_cleanliness"] < 3) & (df["review_scores_rating"] > 3)]
filtro8.to_csv("Ejercicio8.csv")
filtro8
```
