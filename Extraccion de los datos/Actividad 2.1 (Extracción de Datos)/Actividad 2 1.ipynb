**Filtros Basicos para Variables Cualitativas**

**Filtro 1: Por objeto**

Descripción: Se crea un nuevo DataFrame (df) nombrado como filtro1 que contiene únicamente las filas donde la columna NOMBRE_VENDEDOR es igual a "LETICIA RAMIREZ HERNANDEZ".

```python
filtro1=df[(df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
filtro1
# Convertir archivo filtrado a CSV
filtro1.to_csv("F1_PorObjeto.csv")
```

**Filtro 2: Por filas**

Descripción: Se utiliza iloc (index location) para seleccionar filas específicas por su posición, en este caso desde la fila 7 (inclusive) hasta la fila 11 (exclusiva). Los dos puntos (:) indican que se seleccionan todas las columnas de esas filas.

```python
filtro2= df.iloc[7:11, :] # Filas consecutivas
filtro2
filtro2.to_csv("F2_PorFilas.csv")
```

```python
filtro2_1= df.iloc[ [7,10] , :]  #  Filas no consecutivas
filtro2_1
```

**Filtro 3: Por columnas**

Descripción: Se utiliza iloc para seleccionar columnas consecutivas por su posición:
El rango 1:4 indica que se seleccionan las columnas desde la posición 1 (segunda columna) hasta la posición 4 (excluyendo la 4).
El símbolo : antes de la coma indica que se seleccionan todas las filas.

```python
filtro3=df.iloc[ : , 1:4] # Columnas consecutivas 
filtro3
filtro3.to_csv("F3_PorColumnas.csv")
```

Descripción: Se seleccionan columnas no consecutivas del DataFrame:
[1, 3, 7, 10] indica las posiciones de las columnas seleccionadas (segunda, cuarta, octava y undécima columna).
El símbolo : antes de la coma indica que se seleccionan todas las filas.

```python
filtro3_1=df.iloc[ : , [1,3,7,10]] # Columnas no consecutivas  
filtro3_1
```

**Cargar archivo con Indice**

Descripción: Se utiliza el parámetro index_col=0 para especificar que la primera columna del archivo (Detalle precios y productos fabricados 2022.xlsx) será utilizada como índice del DataFrame.

```python
# Indice
df1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=0)
df1
```

**Filtro 4: filas vs. columnas utilizando el índice**

Descripción: Se utiliza loc para filtrar el DataFrame df1 con base en:
Filas específicas: ["F37034", "F39838"] (se seleccionan únicamente estas dos filas según los valores de su índice).
Columnas específicas: ["NOMBRE_CLIENTE"] (se selecciona únicamente esta columna).

```python
# Paso 2: aplicamos filtro de filas vs columnas (unicamente df con indice)

filtro4= df1.loc[["F37034", "F39838"], ["NOMBRE_CLIENTE"]]
filtro4
# filtro4.to_csv("F4_FilasVSColumnas.csv")
```

**Filtro 5: Por cabecera**

Descripción: Utiliza el método head() para seleccionar las primeras n filas del DataFrame.
En este caso, se especifica el número 4 para extraer las primeras 4 filas.

```python
# Filtro por cabecera (se puede especificar el numero)
filtro5= df.head(4)
filtro5
filtro5.to_csv("F5_Cabecera.csv")
```

**Filtro 6: Por cola**

Descripción: Utiliza el método tail() para seleccionar las últimas n filas del DataFrame.
En este caso, se especifica el número 6 para extraer las últimas 6 filas.


```python
# Filtro por cola
filtro6= df.tail(6)
filtro6
filtro6.to_csv("F6_Cola.csv")
```

**Filtros Basicos para Variables Continuas**

**Filtro 7: por Comparación**

Descripción: Este filtro utiliza una condición lógica para seleccionar únicamente las filas donde la columna PRECIO_UNITARIO contiene valores mayores a 20000.
Se aplica directamente sobre el DataFrame df utilizando notación de corchetes ([]), generando un subconjunto de datos que cumple con la condición especificada.

```python
# Filtro por comparación
filtro7= df[df["PRECIO_UNITARIO"] > 20000]
filtro7
filtro7.to_csv("F7_PorComparacion.csv")
```

**Filtros Basicos con Conectores Lógicos**

**Filtro 8: Filtro Y**

Descripción: Este filtro selecciona las filas del DataFrame df que cumplen simultáneamente con dos condiciones:
El valor en la columna PRECIO_UNITARIO es mayor a 20000.
El valor en la columna COSTO_UNITARIO es mayor a 100.
Utiliza el operador lógico & (AND) para combinar ambas condiciones.

```python
filtro8 = df[(df["PRECIO_UNITARIO"] > 20000) & (df["COSTO_UNITARIO"] > 100)]
filtro8
filtro8.to_csv("F8_FiltroY.csv")
```

```python
# Filtro Y (con otro ejemplo)
filtro8_1= df[(df["PRECIO_UNITARIO"] > 2000 ) & (df["FECHA_DOC"] == "2022-01-05")]
filtro8_1
```

**Filtro 9: Filtro O**

Descripción: Este filtro selecciona las filas del DataFrame df que cumplen al menos una de las siguientes condiciones:
El valor en la columna PRECIO_UNITARIO es mayor a 2000.
El valor en la columna FECHA_DOC es igual a "2022-01-05".
Utiliza el operador lógico | (OR) para combinar las condiciones.

```python
# Filtro o (o una o otra)
filtro9= df[(df["PRECIO_UNITARIO"] > 2000 ) | (df["FECHA_DOC"] == "2022-01-05")]
filtro9
filtro9.to_csv("F9_FiltroO.csv")
```

**Filtro 10: Filtro Not**

Descripción: Este filtro selecciona todas las filas del DataFrame df excepto aquellas donde el valor de la columna FECHA_DOC sea igual a "2022-01-05".
Utiliza el operador de negación ~ para invertir el resultado de la comparación df["FECHA_DOC"] == "2022-01-05".


```python
# Filtro not (me da las que no resulten en esta fecha)
filtro10= df[~(df["FECHA_DOC"] == "2022-01-05")]
filtro10
filtro10.to_csv("F10_FiltroNot.csv")
```
